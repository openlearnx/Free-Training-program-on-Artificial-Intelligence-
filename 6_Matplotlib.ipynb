```
y=[5,6,7,9,5,9,8,7,0,6]
```

```
plt.figure(figsize=(9,3))
plt.plot(x,y)
```

```
plt.figure(figsize=(20,7))
plt.plot(x,y)
```

```
plt.figure(figsize=(9,3))
plt.bar(x,y)
```
